# Example of geostrophic kinematic deformation reconstruction 
## from eNATL60-no tide SSH map in Western Mediterranean region

```python
import numpy as np
import pylab as plt
import sys, os 
import xarray as xr 
import warnings 
warnings.filterwarnings('ignore')
```

```python
import pydssa.class_vertical_reconstruction as vr
```

## 1. Download SSH field and depth

```python
path_ssh = '../data/Surf_fields/eNATL60MEDWEST-BLB002_y2009m07d01.1h_sossheig.nc'    
with xr.open_dataset(path_ssh, mask_and_scale=True) as (ds_2):   
        day0 = 0 # hour = 0
        crop = 200
        ssh = np.zeros_like(ds_2.sossheig[day0,crop:-crop,crop:-crop]) 
        ssh[:,:] = ds_2.sossheig[day0,crop:-crop,crop:-crop]
        lon = np.zeros_like(ds_2.nav_lon[crop:-crop,crop:-crop]) 
        lon[:,:] = ds_2.nav_lon[crop:-crop,crop:-crop] 
        lat = np.zeros_like(ds_2.nav_lat[crop:-crop,crop:-crop])
        lat[:,:] = ds_2.nav_lat[crop:-crop,crop:-crop]      
        
path_ssh = '../data/Surf_fields/eNATL60MEDWEST-BLB002_y2009m07d01.1h_sosstsst.nc'    
with xr.open_dataset(path_ssh, mask_and_scale=True) as (ds_2):  
        sst = np.zeros_like(ds_2.sosstsst[day0,crop:-crop,crop:-crop]) 
        sst[:,:] = ds_2.sosstsst[day0,crop:-crop,crop:-crop]  
        
ssh0 = np.nan_to_num(ssh)
sst0 = np.nan_to_num(sst)
lon1 = lon[lon!=0]
lat1 = lat[lat!=0]
my_lon = np.linspace(np.min(lon1),np.max(lon1),np.shape(lon)[1])
my_lat = np.linspace(np.min(lat1),np.max(lat1),np.shape(lat)[0])
lon0,lat0 = np.meshgrid(my_lon,my_lat)



depth = []
with open('../data/NATL60_depth.txt') as fp:
    for cnt, line in enumerate(fp):
        if cnt != 0 and cnt < 109:
            depth.append(np.float(line.strip()))
            
print(np.shape(ssh))
print(np.shape(sst))
```

## 2. Create an ocean reconstruction instance

```python
my_geokindef_reconstruction = vr.vertical_reconstruction()
```

## 3. Enter reconstruction parameters

```python
my_geokindef_reconstruction.ssh = ssh0 
my_geokindef_reconstruction.sst = sst0
my_geokindef_reconstruction.lon = lon0
my_geokindef_reconstruction.lat = lat0                   
```

## 4. Run the geostrophic kinematic deformation reconstruction 

```python
my_geokindef_reconstruction.run_geokindef()
```

## 5. Plot the reconstructed variables

```python
my_geokindef_reconstruction.plot_reconstruction_outputs(crop1 = 100, crop2 = -200, crop3 = 20, crop4 = -20)
```

## 6. Compare reconstructed w to the reference

```python
path_ref = '../data/Reference/NATL60OSM-CJM165_y2013m01d15.1d_W.nc'    
with xr.open_dataset(path_ref, mask_and_scale=True) as (ds_2):   
        w_ref = np.zeros_like(ds_2.vovecrtz[0,0,:,:])    
        w_ref[:,:] = np.mean(ds_2.vovecrtz[0:24,10,:,:] ,0)
        lon_ref = np.zeros_like(ds_2.nav_lon[:,:] )   
        lon_ref[:,:] = ds_2.nav_lon[:,:]   
        lat_ref = np.zeros_like(ds_2.nav_lat[:,:] )   
        lat_ref[:,:] = ds_2.nav_lat[:,:]    
        
w_ref = np.ma.masked_invalid(w_ref)
 
```

```python
path_ref = '../data/Reference/eNATL60MEDBAL-BLB002_y2009m07d01.1h_vovecrtz_0-1000m.nc'    
with xr.open_dataset(path_ref, mask_and_scale=True) as (ds_2):    
        w_ref = np.zeros_like(ds_2.vovecrtz[0,0,:,:])    
        w_ref[:,:] = ds_2.vovecrtz[0,0,:,:] 
        lon_ref = np.zeros_like(ds_2.nav_lon[:,:] )   
        lon_ref[:,:] = ds_2.nav_lon[:,:]   
        lat_ref = np.zeros_like(ds_2.nav_lat[:,:] )   
        lat_ref[:,:] = ds_2.nav_lat[:,:]   
        
w_ref = np.nan_to_num(w_ref)
lon1 = lon_ref[lon_ref!=0]
lat1 = lat_ref[lat_ref!=0]
my_lon = np.linspace(np.min(lon1),np.max(lon1),np.shape(lon_ref)[1])
my_lat = np.linspace(np.min(lat1),np.max(lat1),np.shape(lat_ref)[0])
lon_ref,lat_ref = np.meshgrid(my_lon,my_lat)
        
w_ref = np.ma.masked_invalid(w_ref)
```

```python
# Cropping reconstructed w
crop1 = 160
crop2 = -75
crop3 = 175
crop4 = -180

# Cropping reference w
crop5 = 0
crop6 = -1
crop7 = 0
crop8 = -1

plt.figure(figsize=(35,6))

# Computing levels from Q_tg divergence max
max_var0 = np.max(np.abs(my_geokindef_reconstruction.geokindef[crop5:crop6,crop7:crop8])) 
levs0 = np.linspace(-max_var0*0.01,max_var0*0.01, 51)
levs0 = np.arange(-0.1,0.1, 2*0.1/50)*1e-13
w_reconst = np.copy(my_geokindef_reconstruction.geokindef[:,:])
w_reconst[ssh0==0]=0
w_reconst[w_reconst>0.1e-11]=0
w_reconst[w_reconst<-0.1e-11]=0

plt.subplot(151)
plt.contourf(my_geokindef_reconstruction.lon[crop1:crop2,crop3:crop4],my_geokindef_reconstruction.lat[crop1:crop2,crop3:crop4],w_reconst[crop1:crop2,crop3:crop4],cmap=plt.cm.get_cmap('bwr'),levels=levs0,extend='both')
plt.colorbar() 
plt.ylabel('Geostrophic kinematic deformation',fontsize=18)
plt.title(r'$\nabla_H \cdot Q_{tg}$ at 0 m',fontsize=15) 
 
# Computing levels from w_ref max
max_var = np.max(np.abs(w_ref[crop5:crop6,crop7:crop8]*1e3))
levs = np.linspace(-max_var,max_var, 51) 

plt.subplot(152)
plt.contourf(lon_ref[crop5:crop6,crop7:crop8],lat_ref[crop5:crop6,crop7:crop8],w_ref[crop5:crop6,crop7:crop8]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar() 
plt.ylabel('Reference',fontsize=18)
plt.title('Vertical velocity (10$^{-3}$m/s) at -17 m',fontsize=15) 
 
plt.show()    
```
